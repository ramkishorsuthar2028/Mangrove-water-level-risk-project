```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/mangrove_water_level.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(columns=["Timestamp"],inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df["Land_Use_Type"]=label_encoder.fit_transform(df["Land_Use_Type"])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df["Storm_Event"]=label_encoder.fit_transform(df["Storm_Event"])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df["Flood_Event"]=label_encoder.fit_transform(df["Flood_Event"])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df["Season"]=label_encoder.fit_transform(df["Season"])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df["Risk"]=label_encoder.fit_transform(df["Risk"])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns=["Risk"])
y = df["Risk"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy = accuracy_score(ytest, ypred)
conf_matrix = confusion_matrix(ytest, ypred)
class_report = classification_report(ytest, ypred)

print("Model Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```
